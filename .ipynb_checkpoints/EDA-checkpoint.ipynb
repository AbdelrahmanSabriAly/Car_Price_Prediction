### Data Cleaning
- Missing values
- Duplicated values
- Features' types
- Outliers

```python
# "-" values in Levy column
df['Levy'] = df['Levy'].replace("-",0)
```

```python
df.duplicated().sum()
```

```python
# Duplicated values
df = df.drop_duplicates()
df.shape
```

```python
# Remove rows containing 'Turbo' in the 'Column' column
df = df[~df['Engine volume'].str.contains('Turbo')]
```

```python
# Remove KM from Mileage column
df['Mileage'] = df['Mileage'].apply(lambda x: x.split(' ')[0])
```

```python
df['Doors'].value_counts()
```

```python
#replace 04-may, 02-mar, >5 in door columns with 4,2,5
df['Doors'] = df['Doors'].replace("04-May",4)
df['Doors'] = df['Doors'].replace("02-Mar",2)
#df['Doors'] = df['Doors'].replace(">5",5)
```

```python
df['Doors'].value_counts()
```

```python
# Convert numeric columns types to integers
df['Levy'] = df['Levy'].astype("int")
df['Engine volume'] = df['Engine volume'].astype("float")
df['Mileage'] = df['Mileage'].astype("float")
df.info()
```

```python
df.shape
```

### Feature engineering

```python
df['Age'] = 2023 - df['Prod. year']
df.head()
```

```python
# Prices less than 500 USD are non-realistic
df = df[df['Price']>=500]
df.shape
```

```python
cont_col = ['Price', 'Levy',  'Engine volume', 'Mileage', 'Airbags','Age']
def outlier_handle(data, col_list):
    df_new = data.copy()
    for col_name in col_list:
        q1 = data[col_name].quantile(0.25)
        q3 = data[col_name].quantile(0.75)
        iqr = q3-q1  #IQR
        fence_low  = q1-1.5*iqr
        fence_high = q3+1.5*iqr
        df_new.loc[:,  col_name] = np.where(data[col_name]> fence_high, fence_high,
                                         np.where(data[col_name]< fence_low, fence_low,
                                                  data[col_name]))
    return df_new
df = outlier_handle(df, cont_col)
df.shape
```

```python
def outlier_prcnt(data, col_list):
    for col_name in col_list:
        q1 = data[col_name].quantile(0.25)
        q3 = data[col_name].quantile(0.75)
        iqr = q3-q1  #IQR
        fence_low  = q1-1.5*iqr
        fence_high = q3+1.5*iqr

        
        outliers = ((df[col_name] > fence_high) | (df[col_name] < fence_low)).sum()
        total = df[col_name].shape[0]
        print(f"Total outliers in {col_name} are: {outliers} - {round(100*(outliers)/total,2)}%.")

outlier_prcnt(df, cont_col)
```

### Univariate Analysis
- Distribution of each column

```python
num_col = df.drop(['ID','Prod. year'],axis = 1).select_dtypes(exclude="object").columns
cat_col =  df.select_dtypes(include="object").columns

for col in num_col:
    if (col == 'Airbags')|(col == 'Cylinders'):
        # Sample data
        x = df[col].value_counts()[:10].index
        y = df[col].value_counts()[:10]

        # Create bar plot
        fig = go.Figure(data=go.Bar(x=x, y=y))
        title = "Top 10 "+ col +" number"
        fig.update_layout(title=title,height=400, width=800)


        # Show the plot
        fig.show()
    
    else:
    
        # Create a histogram
        fig = go.Figure(data=[go.Histogram(x=df[col],opacity=0.7,marker=dict(color='#c32148'))])

        # Customize the layout
        fig.update_layout(title=col, xaxis_title='Value', yaxis_title='Frequency',barmode='overlay',
            bargap=0.1,height=400, width=800)

        # Show the chart
        fig.show()
```

### Insights from numeric columns:
- Most levies are between 0 and 24 USD
- 2-litre engine volume is the most common
- Longest mileage occured between 0 and 2500 KM
- Most common cylinder number is 4, then 12
- Most common Airbag number is 4, then 12
- Most frequent car age is between 9 and 11 years

```python
for col in cat_col:
    
    if (col == 'Manufacturer')|(col == 'Model'):
        # Sample data
        x = df[col].value_counts()[:10].index
        y = df[col].value_counts()[:10]

        # Create bar plot
        fig = go.Figure(data=go.Bar(x=x, y=y))
        title = "Top 10 "+ col +'s'
        fig.update_layout(title=title,height=400, width=800)


        # Show the plot
        fig.show()
        
    else:
        labels = df[col].value_counts().index
        values = df[col].value_counts()

        fig = go.Figure(data=[go.Pie(labels=labels, values=values)])

        # Set the title of the chart
        fig.update_layout(title=col,height=600, width=800)

        # Display the chart
        fig.show()
```

### Insight from categorical columns:

- Sonata is the most manufactured car model
- Sedan is the most common car category
- Most cars have leather interior
- Most cars support petrol as a fuel source
- Most cars have an automatic gear box type
- Most cars support front drive wheels
- Most cars have 4 doors
- Most cars have left-wheel
- Most common car colors are black, white and silver
